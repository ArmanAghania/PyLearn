```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balloons_length = np.random.uniform(10, 25, 100)  
balloons_width = np.random.uniform(10, 25, 100)   
balloons_weight = np.random.uniform(5, 50, 100)  
melons_length = np.random.uniform(20, 35, 100)    
melons_width = np.random.uniform(20, 35, 100)    
melons_weight = np.random.uniform(1000, 5000, 100) 

# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(balloons_length, balloons_width, balloons_weight, c='r', marker='o', label='Balloons')
ax.scatter(melons_length, melons_width, melons_weight, c='g', marker='^', label='Melons')

# Labeling
ax.set_xlabel('Length (cm)')
ax.set_ylabel('Width (cm)')
ax.set_zlabel('Weight (g)')
ax.legend()

plt.show()
```
