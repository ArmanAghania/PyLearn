```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

image = cv2.imread('input/wolf.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.show()
```

```python
ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1, cmap='gray')
plt.show()
```

```python
contours, _ = cv2.findContours(thresh1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
result = image.copy()
for contour in contours:
    if cv2.contourArea(contour) > 1000:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x, y), (x + w, y + h), (255), 2)

plt.imshow(result, cmap='gray')

plt.show()
```

```python
image_puma = cv2.imread('input/puma.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image_puma, cmap='gray')
plt.show()
```

```python
image_puma = 255 - image_puma

plt.imshow(image_puma, cmap='gray')

plt.show()
```

```python
contours, _ = cv2.findContours(image_puma, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

result = image_puma.copy()

for contour in contours:
    if cv2.contourArea(contour) > 10:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x, y), (x + w, y + h), (127, 255, 255), 2)

plt.imshow(result, cmap='gray')
plt.show()
```

```python
image_coin = cv2.imread('input/coins.jpg', cv2.IMREAD_GRAYSCALE)
plt.axis('off')

plt.imshow(image_coin, cmap='gray')
plt.show()
```

```python
_, image_coin_thresh = cv2.threshold(image_coin, 100, 255, cv2.THRESH_BINARY)

plt.imshow(image_coin_thresh, cmap='gray')
plt.show()
```

```python
contours, hierarchy = cv2.findContours(image_coin_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
result = image_coin_thresh.copy()
for i, contour in enumerate(contours):
    if hierarchy[0][i][3] == 1:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x, y), (x + w, y + h), (127, 127, 127), 3)

plt.axis('off')
plt.imshow(result, cmap='gray')
plt.show()
```
