```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('captain_tsubasa.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
print(image_gray.shape)
```

```python
cv2.imwrite('captain_tsubasa_gray.jpg', image_gray)
```

```python
image_gray[10:50, 15:65] = 255
plt.imshow(image_gray, cmap='gray')
```

```python
image_gray[0:50, 0:840] = 0
image_gray[0:1200, 790:840] = 0
image_gray[0:1200, 0:50] = 0
image_gray[1150:1200, 0:840] = 0

plt.imshow(image_gray, cmap='gray')
```

```python
image_gray[0:50, 0:768] = 255
image_gray[0:432, 718:768] = 255
image_gray[0:432, 0:50] = 255
image_gray[382:432, 0:768] = 255


plt.imshow(image_gray, cmap='gray')
```

```python
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

tsubasa = image_copy[580:800, 350:550]
```

```python
plt.imshow(tsubasa, cmap='gray')
```

```python
image_gray_copy = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
image_gray_copy[0:220, 330:530] = tsubasa

plt.imshow(image_gray_copy, cmap='gray')
```

```python
cv2.imwrite('captain_tsubasa_gray_copied.jpg', image_gray_copy)
```
